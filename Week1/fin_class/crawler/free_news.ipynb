```python
import sys
import pickle
import requests
from datetime import datetime ,timedelta
from bs4 import BeautifulSoup
```

```python
start_date = "2019-02-27"
stop_date = "2019-03-04"

start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")

date_list = list()
while start <= stop:
    date_list.append(start.strftime('%Y%m%d'))
    start = start + timedelta(days=1)
```

```python
def get_date(date):
    li_date = datetime.strptime(date, "%Y%m%d").strftime('%Y-%m-%d')
    return(li_date)
    
def get_title(news_block_node):
    li_title = news_block_node.select_one('p').get_text()
    return li_title

def get_link(news_block_node):
    li_link = 'http://news.ltn.com.tw/' + news_block_node.select_one('a')['href']
    return li_link

def get_content(li_link):
    r = requests.get(li_link)
    soup = BeautifulSoup(r.text, 'html.parser')
    article_node = soup.select('div.text >p')
    content = ''
    for p in article_node:
            if not 'appE1121' in p.get('class', []):
                content += p.get_text()
    return content
```

```python
def get_news_info(each_news,date):
    date  = get_date(date)
    title = get_title(each_news)
    link  = get_link(each_news)
    content = get_content(link)
    
    info = {'date' : date,
            'title': title,
            'link' : link,
            'content': content}
    return(info)

def get_page_news(page_url,date):
    r = requests.get(page_url)
    r.encoding = "UTF-8"

    soup = BeautifulSoup(r.text, 'html.parser')
    news_blocks = soup.select('ul.list > li')
    
    news = []
    for each_news in news_blocks:
        try:
            news_info = get_news_info(each_news,date)
            print(get_title(each_news))
            news.append(news_info)
        except:
#             print('-------{}-------'.format())
            pass
        #news.append(news_info)
    return(news)
```

```python
def get_new_talk_news(date_list, url="https://news.ltn.com.tw/list/newspaper/politics/"):
    data = []
    for date in date_list:
        print("date: {}".format(date))
        data = data + get_page_news( url+str(date),date )
        
    print('done')
    return(data)
```

```python
data = get_new_talk_news(date_list)
```
