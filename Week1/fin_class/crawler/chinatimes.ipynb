```python
import sys
import pickle
import requests
from datetime import datetime
from bs4 import BeautifulSoup
```

## enter the url of the home of news

```python
def get_date(news_block_node):
    date_string = news_block_node.find(class_="date").string
    return(datetime.strptime(date_string, '%Y/%m/%d').strftime('%Y-%m-%d'))
    
def get_title(news_block_node):
    return news_block_node.find(class_="cropper").a.get('title')

def get_link(news_block_node):
    return 'http:' + news_block_node.find(class_="cropper").a.get('href')

def get_content(link):
    r = requests.get(link)
    r.encoding = "UTF-8"
    soup = BeautifulSoup(r.text, 'html.parser')
    article_node = soup.select_one("article").select("p")
    content = ''
    for p in article_node :
        content += p.text
    return content.replace("\n", "")
```

```python
def get_news_info(each_news):
    date  = get_date(each_news)
    title = get_title(each_news)
    link  = get_link(each_news)
    content = get_content(link)
    
    info = {'date' : date,
            'title': title,
            'link' : link,
            'content': content}
    return(info)

def get_page_news(page_url):
    r = requests.get(page_url)
    r.encoding = "UTF-8"

    soup = BeautifulSoup(r.text, 'html.parser')
    news_blocks = soup.find_all(class_= "clear-fix")
    
    news = []
    for each_news in news_blocks:
        try:
            news_info = get_news_info(each_news)
            print(get_title(each_news))
            news.append(news_info)
        except:
#             print('-------{}-------'.format())
            pass
        #news.append(news_info)
    return(news)


def get_new_talk_news(from_page=1, end_page=5, url="https://www.chinatimes.com/politic/total/"):
    print("page_number from {} to {}".format(from_page, end_page -1))
    data = []
    for page_number in range(from_page, end_page):
        print("page_number: {}".format(page_number))
        if page_number >= 2 :
#            r = requests.get(url+'?page='+str(page_number) )
#            r.encoding = "UTF-8"    
#            soup = BeautifulSoup(r.text, 'html.parser')
#            news_blocks = soup.find_all(class_= "clear-fix")  
#            news = []
#            for each_news in news_blocks:
#                    try:
#                        info = get_news_info(each_news)
#                        print(get_title(each_news))
#                    except:
#             print('-------{}-------'.format())
#                       pass   
#            news.append(info)
#            data = data + news
            data = data + get_page_news( url+'?page='+str(page_number) )
        else : 
            data = data + get_page_news( url )
#            r = requests.get(url )
#            r.encoding = "UTF-8"    
#            soup = BeautifulSoup(r.text, 'html.parser')
#            news_blocks = soup.find_all(class_= "clear-fix")  
#            news = []
#            for each_news in news_blocks:
#                    try:
#                        info = get_news_info(each_news)
#                        print(get_title(each_news))
#                    except:
#             print('-------{}-------'.format())
 #                       pass   
 #           news.append(info)
 #           data = data + news
    print('done')
    return(data)
```

### 刪除重複文章  此方法爬中國時報會出現首頁的第一篇重複出現三次的情況

```python
length = len(data) 
x=0
while x < length-1:
    if data[x+1] == data[x]:
        data.remove(data[x])
        x -= 1
        length -= 1
    x +=1
```
